### Question 1: Correlation vs. Causation

**Answer:**
Snow isolated variables by comparing affected vs unaffected groups. Brewery workers near the pump didn't get sick (they drank beer). The Hampstead widow far away got sick (she drank pump water). He removed the pump handle and deaths stopped. This proved causation, not just correlation.

### Question 2: The Danger of Missing Data

**Answer:**
Hospital records only capture severe cases from wealthier people. Poor people who died at home wouldn't be recorded. The geographic map would be wrong (showing hospital locations, not actual deaths). This selection bias would hide the true pattern around Broad Street and lead to wrong conclusions.

**Answer:**
- **AI = Efficiency**: Assumed true, but implementation costs often outweigh gains
- **More Data = Better**: False data quality matters more than quantity
- **Expensive = Better**: Often false—inefficiency, not cost, is the problem
